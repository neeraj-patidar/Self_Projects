```python
#Live Sketcher

import cv2
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

def imshow(name, img):
  cv2_imshow(img)
def sketch(ime):
    gray_img = cv2.cvtColor(ime, cv2.COLOR_BGR2GRAY)
    im_g_blur = cv2.GaussianBlur(gray_img,(5,5),0)
    
    canny_edges = cv2.Canny(im_g_blur,10,70)
    ret,sk = cv2.threshold(canny_edges,70,255,cv2.THRESH_BINARY_INV)
    return sk

cap = cv2.VideoCapture(0)

while True:
    #
    ret,frame = cap.read()
    imshow('Our Live Sketcher',sketch(frame) )
    if cv2.waitKey(1) == 13:
        break

cap.release()
cv2.destroyAllWindows()
```
